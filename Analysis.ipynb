## Data Exploration begins!

### Q1. Which was the best month considering Sales?

```python
combined['Sale'] = combined['Quantity Ordered']*combined['Price Each']
```

```python
combined['Order Month'] = combined['Order Date'].dt.month
```

```python
combined
```

```python
bar1 = combined.groupby(['Order Month'])[['Sale']].agg('sum')
bar1
```

```python
combined.groupby(['Order Month'])[['Sale']].agg('sum').sort_values('Sale',ascending = False).head(1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(bar1.index,bar1['Sale'])
plt.xticks(bar1.index)
```

### Q2. Which city produced the maximum sales?

```python
def get_city(x):
    return x.split(',')[1].strip()
def get_state(x):
    return x.split(',')[2].split()[0].strip()
```

```python
combined.columns
```

```python
combined['City'] = combined['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
```

```python
combined
```

```python
bar2 = combined.groupby(['City'])[['Sale', 'Quantity Ordered']].agg('sum')
bar2
```

```python
combined.groupby(['City'])[['Sale', 'Quantity Ordered']].agg('sum').sort_values('Sale',ascending = False).head(1)
```

```python
plt.bar(bar2.index,bar2['Sale'])
plt.xticks(bar2.index,rotation='vertical')
```

```python
fig, ax = plt.subplots()
ax.bar(bar2.index,bar2['Sale'])

# Get current ticks and format them
yticks = ax.get_yticks()
formatted_yticks = [int(tick) for tick in yticks]  # Format ticks with commas

# Apply the formatted labels
ax.set_yticklabels(formatted_yticks)
ax.set_xticklabels(bar2.index,rotation='vertical')
```

### Q4. Which products are most often bought together?

```python
def get_list(x):
    glist = []
    for y in x['Product']:
        glist.append(y)
    return glist
```

```python
orders = combined.groupby(['Order ID']).apply(get_list) # see in notes how groupby is comprehended
```

```python
orders.name = 'items'
```

```python
orders
```

```python
from itertools import combinations
from collections import Counter
```

```python
pairs = []
for items in orders:
    pairs.extend(combinations(sorted(items),2)) # sorted to get ['a','b'] and ['b','a'] same
```

```python
Counter(pairs)
```

```python
dict = Counter(pairs)
```

```python
pd.DataFrame(dict.items(),columns=['Pair','Freq']).sort_values('Freq',ascending=False).head(10)
```

### Q5. Which product sold the most?

```python
bar3 = combined.groupby(['Product'])['Quantity Ordered'].agg('sum')
bar3
```

```python
plt.bar(bar3.index, bar3.values)
plt.ylabel('Quantity Sold')
plt.xlabel('Product')
plt.xticks(bar3.index, rotation='vertical',size=8)
plt.show() #no xticks stuff
```

### Hypothesized that cheaper products may have a greater count, let's check:

```python
line2 = combined.groupby(['Product'])['Price Each'].agg('mean')
line2
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(bar3.index, bar3.values, color='g')
ax2.plot(bar3.index, line2.values, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(bar3.index, rotation='vertical', size=8)

fig.show()
```

### Saw the correlation between the quantity sold and price of that product in this graph, and yes indeed it
